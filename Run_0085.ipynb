```python
#imports

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import scipy as sp
import h5py

import pytpc

from mpl_toolkits.mplot3d import Axes3D

```

```python
full = pytpc.HDFDataFile('/home/chen/run_0085.h5', 'r') 
```

```python
print(len(full))
```

```python
i = 228
evt_full = full[i]
xyzs = evt_full.xyzs(peaks_only=True, return_pads=True, baseline_correction=True, cg_times=True)
```

```python

fig, ax = plt.subplots(1, 2, figsize=(18, 9))

ax[0].plot(xyzs[:,0], xyzs[:,1], '.')
ax[0].axis('equal')
ax[0].set_xlabel('x [mm]')
ax[0].set_ylabel('y [mm]')
ax[0].set_title('xy')
ax[0].set_xlim((-275, 275))
ax[0].set_ylim((-275, 275))


ax[1].plot(xyzs[:,2], xyzs[:,1], '.')
ax[1].set_xlabel('z [mm]')
ax[1].set_ylabel('y [mm]')
ax[1].set_title('zy')
ax[1].set_xlim((0, 2000))
ax[1].set_ylim((-275, 400))

fig2 = plt.figure(figsize=(16,16))
ax2 = fig2.add_subplot(111, projection="3d")
ax2.scatter(xyzs[:,0], xyzs[:,1], xyzs[:,2])
ax2.axis('equal')
ax2.set_xlim(-275., 275.)
ax2.set_ylim(-275., 275.)
ax2.set_zlim(0., 512.)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

#plt.savefig('/home/chen/Run_0085.pdf')
```
